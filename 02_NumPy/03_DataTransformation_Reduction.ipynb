# Esercizio 3 — Data Transformation e Data Reduction

**Obiettivo:** applicare tecniche di trasformazione e riduzione dei dati per ottimizzare il dataset.  
**Argomenti:** encoding, scaling, PCA (Principal Component Analysis).


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.decomposition import PCA

print("Librerie importate correttamente")

#creiamo un piccolo dataset e cerchiamo di applicare tecniche di trasformazione e riduzioni di dati 
np.random.seed(42)

città = np.random.choice(["Roma","Milano","Napoli","Palermo"], 10)
età = np.random.randint(20,60,10)
reddito = np.random.randint(20000,70000,10)

df = pd.DataFrame({"Età" : età, "Reddito" : reddito, "Città" : città})

df

```

```python
#Encoding delle variabili categoriche 
# l'obiettivo e' portare le variabili categoriche in numeriche

#Utilizzeremo from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoded_values = encoder.fit_transform(df["Città"])

#in questo modo ad ogni citta' sara' assegnato un valore numerico 

print("", encoded_values) #Mostro la colonna trasformata

print("", encoder.classes_) #Mostro il criterio di trasformazione

df["Encoded_values"] = encoded_values
df
    
```

```python
#One-Hot encoding
# Con questo metodo di data porting creiamo una colonna binaria per ogni categoria
# usiamo sempre sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoded_values = encoder.fit_transform(df[["Città"]])

#ora encoded values è una matrice binaria

#capiamo quali sono le colonne
print(encoder.get_feature_names_out(["Città"]))

#creiamo il database espanso
encoded_df = pd.DataFrame(encoded_values, columns=encoder.get_feature_names_out(["Città"]))
df_encoded = pd.concat([df, encoded_df], axis=1)
df_encoded
```

```python
#Dopo aver effettuato la poratbilita' delle variabili categoriche bisogna mettere tutte le variabili
#sullo stesso ordine di grandezza ==> normalizzare

from sklearn.preprocessing import MinMaxScaler

#Dobbiamo eliminare la colonna originale delle citta'
df_numeric = df_encoded.drop('Città', axis=1)

scaler=MinMaxScaler()
scaled = scaler.fit_transform(df_numeric)

df_scaled = pd.DataFrame(scaled, columns = df_numeric.columns)
df_scaled.describe()
```

```python
#Confronto con StandardScaler
from sklearn.preprocessing import StandardScaler

scaler_std = StandardScaler()
scaled_std = scaler_std.fit_transform(df_numeric)

df_scaled_std = pd.DataFrame(scaled_std, columns = df_numeric.columns)
print(df_scaled_std.head())
```

```python
#Visualizzazione grafica

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("MinMaxScaler [0,1]")
plt.boxplot(df_scaled, vert=False)
plt.subplot(1,2,2)
plt.title("StandardScaler (Z-score)")
plt.boxplot(df_scaled_std, vert=False)
plt.show()
```
